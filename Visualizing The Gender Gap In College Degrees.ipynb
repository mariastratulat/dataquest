## Visualizing The Gender Gap In College Degrees
### Introduction
The aim of this project is to study generating beautiful line charts to compare the gender gap between all degree categories.

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv') # read data
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(left=False, bottom=False)
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
plt.show()
```

All college degree categories in STEM present gender predominance which vary over time. The two which represent a complete change in gender predominance are Psychology in the 1970s which changed from male to female predominance, and Biology in the 1980s also from male to female.

### Setting y-axis label

The trade-off we made is that it's now more difficult for the viewer to discern approximately which years some interesting changes in trends may have happened. This is acceptable because we're primarily interested in enabling the viewer to quickly get a high level understanding of which degrees are prone to gender imbalance and how that has changed over time.

In the vein of reducing cluttering, let's also simplify the y-axis labels.

```python
fig = plt.figure(figsize=(16, 22))

# A function to set some visualization preferences
def chart_vis(cat_list, cat_idx):
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.set_title(cat_list[cat_idx])
    ax.tick_params(bottom=False, left=False, labelbottom=False)

#Generate the first column
for sp in range(0,18,3):
    cat_index = int(sp/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[cat_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[cat_index]], c=cb_orange, label='Men', linewidth=3)
    chart_vis(stem_cats, cat_index)
    if cat_index == 0:
        ax.text(2005, 87, "Men")
        ax.text(2002, 8, "Women")
    elif cat_index == 5:
        ax.text(2005, 62, "Men")
        ax.text(2001, 35, "Women")
        ax.tick_params(labelbottom="on")
        
#Generate the secont column
for sp in range(0,15,3):
    cat_index = int(sp/3)
    ax = fig.add_subplot(6,3,sp+2)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[cat_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[cat_index]], c=cb_orange, label='Men', linewidth=3)
    chart_vis(lib_arts_cats, cat_index)
    if cat_index == 0:
        ax.text(2005, 87, "Men")
        ax.text(2002, 8, "Women")
    elif cat_index == 4:
        ax.tick_params(labelbottom="on")
        
#Generate the third column
for sp in range(0,18,3):
    cat_index = int(sp/3)
    ax = fig.add_subplot(6,3,sp+3)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[cat_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[cat_index]], c=cb_orange, label='Men', linewidth=3)
    chart_vis(other_cats, cat_index)
    if cat_index == 0:
        ax.text(2005, 87, "Men")
        ax.text(2002, 8, "Women")
    elif cat_index == 5:
        ax.text(2005, 62, "Men")
        ax.text(2001, 35, "Women")
        ax.tick_params(labelbottom="on")
```

### Adding a horizontal line

While removing most of the y-axis labels definitely reduced clutter, it also made it hard to understand which degrees have close to 50-50 gender breakdown. While keeping all of the y-axis labels would have made it easier, we can actually do one better and use a horizontal line across all of the line charts where the y-axis label 50 would have been.

```python
fig = plt.figure(figsize=(16, 22))

# A function to set some visualization preferences
def chart_vis(cat_list, cat_idx):
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.set_title(cat_list[cat_idx])
    ax.tick_params(bottom=False, left=False, labelbottom=False)
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    
#Generate the first column
for sp in range(0,18,3):
    cat_index = int(sp/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[cat_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[cat_index]], c=cb_orange, label='Men', linewidth=3)
    chart_vis(stem_cats, cat_index)
    if cat_index == 0:
        ax.text(2005, 87, "Men")
        ax.text(2002, 8, "Women")
    elif cat_index == 5:
        ax.text(2005, 62, "Men")
        ax.text(2001, 35, "Women")
        ax.tick_params(labelbottom="on")
        
#Generate the secont column
for sp in range(0,15,3):
    cat_index = int(sp/3)
    ax = fig.add_subplot(6,3,sp+2)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[cat_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[cat_index]], c=cb_orange, label='Men', linewidth=3)
    chart_vis(lib_arts_cats, cat_index)
    if cat_index == 0:
        ax.text(2005, 87, "Men")
        ax.text(2002, 8, "Women")
    elif cat_index == 4:
        ax.tick_params(labelbottom="on")
        
#Generate the third column
for sp in range(0,18,3):
    cat_index = int(sp/3)
    ax = fig.add_subplot(6,3,sp+3)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[cat_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[cat_index]], c=cb_orange, label='Men', linewidth=3)
    chart_vis(other_cats, cat_index)
    if cat_index == 0:
        ax.text(2005, 87, "Men")
        ax.text(2002, 8, "Women")
    elif cat_index == 5:
        ax.text(2005, 62, "Men")
        ax.text(2001, 35, "Women")
        ax.tick_params(labelbottom="on")
```
